###Q14: Write a program to find the distance between two points in 3D space. Take all coordinates as input from the user.

###Q18: Given the first term and common difference of an Arithmetic Series, find the sum of first n terms. Take all inputs from the user.

````python
# Write your code here
first_term = int(input("Enter the first number: "))
common_difference = int(input("Enter the common difference: "))
n = int(input("Enter the number of terms: "))
sum = n/2*(2*first_term + (n - 1)*common_difference)
print("Sum of first n terms: ",sum)
````
