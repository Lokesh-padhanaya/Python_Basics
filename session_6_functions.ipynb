# Let's create a function(with docstring)

```python
def is_even(num):
  """
  This function returns if a number is odd or even
  input - any valid integer
  output - odd/even
  created on - 18-05-2026
  """
  if type(num) == int:
    if num % 2 == 0:
      return 'even'
    else:
      return 'odd'
  else:
    print("please enter the integer data type")
```

## *args and **kwargs
#### *args and **kwargs are special Python keywords that are used to pass the variable length of arguments to a function

```python
# *args
# allows us to pass a variable number of non-keyword arguments to a function.

def multiply(*args):
  product = 1

  for i in args:
    product = product * i
  return product
```

```python
multiply(1,2,3)
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.

def display(**kwargs):

  for (key,value) in kwargs.items():
    print(key,'->', value)

```

```python
display(India='delhi', nepal='kathmandu', srilanka = 'colombo')
```

## Lambda Function
#### A lambda function is a small anonymous function.
#### A lambda function can take any number of arguments, but can only have one expression.


####  **lambda**  a,b **:**  **a + b**
####    keyword,  Parameters,  Expression

```python
# x -> x^2

a = lambda x: x**2
a(2)
```

```python
# x,y -> x + y
b = lambda x,y: x+y
b(2,5)
```

### Difference between lambda vs Normal Function
- No name
- lambda has no return value(infact, returns a function)
- lambda is written in 1 line
- not reusable

### Then why use lambda functions?
#### **They are used with HOF(Higher Order Functions)**

```python
# check if a string has 'a'

a = lambda s: 'a' in s
a('hello')
```

```python
# odd or even

a = lambda x: 'even' if x%2 == 0 else 'odd'
a(5)
```

## Higher Order Functions
- When a function takes one function in input or return the function in output

```python
# Example

def square(x):
  return x**2

def transform(f,L):
  output = []
  for i in L:
    output.append(f(i))
  print(output)

L = [1,2,3,4,5]

transform(square,L)
```

```python
# Butif we want to change the function from square to cube, we will have to agai write a function for cube,

def square(x):
  return x**3

def transform(f,L):
  output = []
  for i in L:
    output.append(f(i))
  print(output)

L = [1,2,3,4,5]

transform(square,L)
```

```python
# But rather than writing a code time and again to change the transformation, we can use the lambda function.

def transform(f,L):
  output = []
  for i in L:
    output.append(f(i))
  print(output)

L = [1,2,3,4,5]

transform(lambda x: x**2,L)
```

### 3 Higher Order Functions

###1.  Map

```python
# square the items of a list
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# odd/even labelling of list itmes
L = [1,2,3,4,5]
list(map(lambda x:'even' if x%2 == 0 else 'odd',L))
```

```python
# fetch names from a list of dict

users = [
    {
        'name': 'Rahul',
        'age':45,
        'gender': 'male'
    },
    {
        'name': 'Lokesh',
        'age': 28,
        'gender': 'male'
    },
    {
        'name': 'Ankita',
        'age': 50,
        'gender': 'female'
    }
]
```

```python
list(map(lambda users: users['name'],users))
```

### 2. Filter

```python
# number greater than 5
L = [1,2,3,4,5,6,7]
list(filter(lambda x : x > 5,L))
```

```python
# fetching fruits starting with 'a'
fruits = ['apple','guava','cherry']
list(filter(lambda x: x.startswith('a'),fruits))
```

### 3. Reduce

```python
# sum of all item
import functools

functools.reduce(lambda x,y: x+y,[1,2,3,4,5])
```

```python
# find min
functools.reduce(lambda x,y: x if x < y else y,[1,2,3,4,5] )
```
