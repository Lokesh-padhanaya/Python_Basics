### A user can find out the distance between 2 coordinates[link text]

```python
class Point:
  def __init__(self,x,y):
    self.x_cor = x
    self.y_cor = y

  def __str__(self):
    return "<{},{}>".format(self.x_cor,self.y_cor)

  def euclidean_distance(self,other):
    return ((self.x_cor - other.x_cor)**2 + (self.y_cor - other.y_cor)**2)**0.5
```

```python
p1 = Point(0,0)
p2 = Point(1,1)
p1.euclidean_distance(p2)
```

### A user can find find the distance of a coordinate from origin



```python
class Point:
  def __init__(self,x,y):
    self.x_cor = x
    self.y_cor = y

  def __str__(self):
    return "<{},{}>".format(self.x_cor,self.y_cor)

  def euclidean_distance(self,other):
    return ((self.x_cor - other.x_cor)**2 + (self.y_cor - other.y_cor)**2)**0.5

  def distance_from_origin(self):
    return ((self.x_cor)**2 + (self.y_cor)**2)**0.5
```

```python
p1 = Point(0,0)
p2 = Point(1,1)
p2.distance_from_origin()
```

### A user can check if a point lies on a given line


```python
class Point:
  def __init__(self,x,y):
    self.x_cor = x
    self.y_cor = y

  def __str__(self):
    return "<{},{}>".format(self.x_cor,self.y_cor)

  def euclidean_distance(self,other):
    return ((self.x_cor - other.x_cor)**2 + (self.y_cor - other.y_cor)**2)**0.5

  def distance_from_origin(self):
    return ((self.x_cor)**2 + (self.y_cor)**2)**0.5

    # for line we have to create a new class
class Line:

  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return '{}x + {}y + {} = 0'.format(self.A, self.B, self.C)

  def point_on_line(line,point):
    if line.A*point.x_cor +line.B*point.y_cor +line.C == 0:
      return "lies in the line"
    else:
      return "does not lie on the line"
```

```python
l1 = Line(3,4,5)
print(l1)

p1 = Point(1,2)
print(p1)

l1.point_on_line(p1)
```

###A user can find the distance between a given 2D point and a given line

```python
class Point:
  def __init__(self,x,y):
    self.x_cor = x
    self.y_cor = y

  def __str__(self):
    return "<{},{}>".format(self.x_cor,self.y_cor)

  def euclidean_distance(self,other):
    return ((self.x_cor - other.x_cor)**2 + (self.y_cor - other.y_cor)**2)**0.5

  def distance_from_origin(self):
    return ((self.x_cor)**2 + (self.y_cor)**2)**0.5

    # for line we have to create a new class
class Line:

  def __init__(self,A,B,C):
    self.A = A
    self.B = B
    self.C = C

  def __str__(self):
    return '{}x + {}y + {} = 0'.format(self.A, self.B, self.C)

  def point_on_line(line,point):
    if line.A*point.x_cor +line.B*point.y_cor +line.C == 0:
      return "lies in the line"
    else:
      return "does not lie on the line"
  def shortest_distance(line,point):
    return abs(line.A*point.x_cor + line.B*point.y_cor + line.C)/(line.A**2 + line.B**2)**0.5
```

```python
l1 = Line(1,1,-2)
p1 = Point(1,1)
l1.shortest_distance(p1)
```

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
# Object without a reference
class Person:

  def __init__(self):
    self.name = "Lokesh"
    self.gender = 'male'

p = Person()
q = p
```

```python
# Multiple ref
print(id(p))
print(id(q))   # both reference variables are pointing to the same position.

# p and q are not the objects while they are the reference variables of the object, "Person()" this is the object which is being stored in p and q.
```

```python
# change attributes value with the help of 2nd object

print(p.name)
print(q.name)
q.name = 'Ankit'
print(p.name)
print(q.name)

# change in one variable made the changes in both variables because both variables are refering to the same object.
```

##Static Variables (Vs Instance Variables)

- value of instance variable is different for every object and value of Static variable is same for every object.
- static variable -> class
- instance variable -> object
- instance variable is declared -> inside the constructor Written as -> object(self).static_variable
- static variable is declared -> outside all the methods but inside the class. Written as -> class.static_variable

##### to access the static method, we dont need to create an object, we can directly access the static method using the class.   Atm.get_cus_id()
