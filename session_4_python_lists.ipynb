# What are Lists?


List is a data type where you can store multiple items under 1 name. More technically, act like **dynamic arrays** which means you can add more items on the fly.


- Why Lists are required in programming ?

# Accessing items to a List

```python
# Indexing
L = [1,2,3,[4,5]]

# positive indexing
print(L[0])

# Negative Indexing
print(L[-3])

# Extracting the 4 form 2D List
print(L[-1][-2])

#or
print(L[3][0])
```

```python
# Indexing in 3D
L = ([[[1,2],[3,4]],[[1,4],[3,5]]])

print(L[0][0][1])
```

# Slicing

```python
L = [1,2,3,4,5,6]

print(L[-3:])

print(L[-5:-2:2])

print(L[::-1])
```

## Adding items to a list

```python
# append (to add one item at a time, at last position)

L= [1,2,3,4,5]

L.append(6)
print(L)

M =[1,2,3,4,5]
M.append([7,8,9])
print(M)
```

```python
# extend (to add multiple items at a time, at last position)
L = [1,2,3,4,5]

L.extend([7,8,9])
print(L)

M =[1,2,3,4,5]
M.extend('delhi')
print(M)
```

```python
# insert()  (inserting the value at my desired position)
L = [1,2,3,4,5]

L.insert(1,100)
print(L)

```

## Editing items in a List (because lists are mutable)

```python
L = [1,2,3,4,5]

# editing with indexing
L[-1] = 500
print(L)

# Editing with slicing
L[1:4] = [200,300,400]
print(L)
```

## Deleting items from a List

```python
# del

L = [1,2,3,4,5]
print(L)

#Indexing
del L[-1]
print(L)

# slicing
del L[1:3]
print(L)

```

```python
# remove

L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
#pop

L = [1,2,3,4,5]
L.pop(0)
print(L)

L.pop()  # default index is last element(-1).
print(L)
```

```python
# clear

L = [1,2,3,4,5]
L.clear()
print(L)
```

# List Comprehension
### List Compreshension provides a concise way of creating lists.
### newlist = [expression for item in iterable if condition == True]


### newlist = [`expression` **for** `item` **in** `iterable` **if** `condition` == **True**]

### Advantages of List Comprehension
- More time-efficient and space-efficient than loops.
- Require fewer lines of code.
- Transforms iterative statement into a formula.



```python
# generally what we would have done.

L = []

for i in range(1,11):
  L.append(i)
print(L)
```

```python
# But according to List comprehenssion.

L = [i for i in range(1,11)]
print(L)
```

```python
# Scalar multiplication on a vector
v = [2,3,4]
s = -3

x = []
for i in v:
  x.append(s*i)
print(x)

# but more effiecient way using List comprehension.

[s*i for i in v]
print(L)
```

```python
# Add squares

L = [1,2,3,4,5]

[i**2 for i in L]

```

```python
# Print all numbers divisible by 5 in the range if 1 to 50.
[i for i in range(1,50) if i%5 == 0]
```

```python
# find languages which start with letter p

languages = ['java', 'python', 'php', 'C', 'javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple', 'guava', 'cherry','banana']
my_fruits = ['apple', 'kiwi', 'grapes','banana']

# add new list from mu_fruits and items if the fruits exists in basket and also starts with 'a'.
```

```python
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith("a")]
```

```python
# Print a (3,3) matrix using list comprehension  ->  Nested List comprehension

[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# Cartesian products -> List comprehension on 2 lists together.

L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

# Zip
### The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

### If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```

# Disadvantages of python Lists

- Slow
- Risky usage
- eats up more memory

```python
# Risky Usage
a = [1,2,3]
b = a
print(a)
print(b)

a.append(4)
print(a)
print(b)

# lists are mutable therefore changes in a also leads the changes in b.

# therefore to avoid this problem, we can use .copy() function.

a = [1,2,3]
b = a.copy()
print(a)
print(b)

a.append(4)
print(a)
print(b)
```

```python
# Create 2 lists from a given list where
# 1st list will contain all the odd numbers from the original list and
# the 2nd one will contain all the even numbers

L = [1,2,3,4,5,6]

list_even = [i for i in L if i%2 == 0]
print(list_even)

list_odd = [j for j in L if j%2 != 0]
print(list_odd)
```
