# Arithmatic Operators

```python
# Addition
print(5+6)

# Subtraction
print(5-6)

# Multiplication
print(5*6)

#Division
print(6/5)

#Integer Division
print(6//5)

# Modulus Operator
print(6%5)

# Powerof Operator
print(5**2)
```

# Relational Operators

```python
print(4<5)

print(4>5)

print(11>=5)

print(15<=20)

print(11==11)

print(4!=5)
```

# Logical Operator

```python
print(1 and 0)

print(1 or 0)

print(not 1)

```

# Bitwise Operators

```python
# Bitwise "and"
print(2 & 3)

# because               2 -> 010
#                       3 -> 110
# after applying "and"->     010 -> 2




# Bitwise "or"

print(2 | 3)

#               2 -> 010
#               3 -> 110
#  bitwise "or"      110 -> 3


# Bitwise "xor"

print(2^3)

# same bit -> 0
# different bit -> 1
#    2 -> 10
#    3 -> 11
# "xor"   01 -> 1


# Bitwise "not"

print(~3)


# leftshift
print(4 >> 2)

# rightshift

print(5 << 2)
```

# If-else in Python

##### If else statement is used in Branching

```python
 # login program and indentation
 # email -> lokeshyadav@gmail.com
 # password -> 1234

 #if condition:
    #code
# else:
    # code

email = input("Enter the email address: ")
password = input("Enter the password: ")


if email == 'lokeshyadav@gmail.com' and password == '1234':
  print("Welcome Lokesh")
else:
  print("Incorrect login ID or password")
```

## "elif"  and  "nested if-else"

```python
email = input("Enter the email address: ")
password = input("Enter the password: ")

if email == "lokeshyadav@gmail.com" and password == "1234":
  print("Welcome Lokesh")

elif email ==  "lokeshyadav@gmail.com" and password != "1234":
  print("Incorrect password")
  password = input("Enter the password again: ")

  if password == "1234":
    print("Welcome, Finally")

  else:
    print("Your ID is blocked for next 1 minute due to 2 time incorrect password")

else:
  print("Incorrect login ID")
```

```python
# minimum of three numbers

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

if a < b and a < c:
  print("minimum number is",a)

elif b < c:
  print("minimum number is",b)

else:
  print("minimum number is",c)
```

```python
# While loop example -> program to print the table

number= int(input("Enter the number: "))

i = 1

while i < 11:
  print(number, '*', i, '=', number * i)
  i+=1

# (We can visualize the code using a tool name: "pythontutor")
```

```python
# while loop with else

x = 1

while x < 5:
  print(x)
  x += 1

else:
  print("limit crossed")
```

```python
# Guessing game

# Generate a random integer between 1 and 100

import random
jackpot = random.randint(1,100)

guess = int(input("Guess the number: "))
counter = 1

while guess != jackpot:
  if guess < jackpot:
    print("Wrong guess! Guess higher")
  else:
    print("wrong guess! Guess lower")
  guess = int(input("Guess the number: "))
  counter += 1

else:
  print("Correct guess")
  print("Attempts",counter)


```
