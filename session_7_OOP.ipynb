```python
# so we used this step, rather than printing anything we called the "menu" function.

class Atm:

  # constructor(special function) -> constructor executes itself, when we create an object of that class. we dont have to explicitly execute this.
  def __init__(self):
    self.pin = ''
    self.balance = 0
    self.menu()

  def menu(self):
    user_input = input("""
    Hi how can i help you.
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    """)

    if user_input == '1':
      #create pin
      self.create_pin()
    elif user_input == '2':
      # change pin
      self.change_pin()
    elif user_input == '3':
      # check balance
      self.check_balance()
    elif user_input == '4':
      # withdraw
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input("Enter the new Pin")
    self.pin = user_pin

    user_balance = int(input("Enter your balance"))
    self.balance = user_balance

    print("pin changed successfully")
    self.menu()

  def change_pin(self):
    old_pin = input("Enter old pin")

    if old_pin == self.pin:
      # let him change the pin
      new_pin = input("enter new pin")
      self.pin = new_pin
      print("Pin changed successfully")
      self.menu()
    else:
      print("you have entered a wrong pin")
      self.menu()

  def check_balance(self):
    user_pin = input("enter your pin")

    if user_pin == self.pin:
      print("your balance is ", self.balance)
    else:
      print("you have entered a wrong pin, so you are a fraudulent")

  def withdraw(self):
    user_pin = input("Enter your pin")
    if user_pin == self.pin:
      # allow to withdraw
      amount = int(input("enter the amount"))
      if amount <= self.balance:
        self.balance = self.balance - amount
        print("withdrawl successful. Your current balance is", self.balance)
      else:
        print("want to become rich, without earning")
    else:
      print("are you a thief ?")
    self.menu()


```

```python
class Fraction:

  # Parameterized constructor
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  # logic for addition
  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den * other.den

    return '{}/{}'.format(new_num,new_den)
```

```python
# similarly for minus, multiplication and division

class Fraction:

  # Parameterized constructor
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  # logic for addition
  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den * other.den

    return '{}/{}'.format(new_num,new_den)

  # subtraction logic
  def   __sub__(self,other):
    new_num = self.num*other.den - self.den*other.num
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  # multiplication

  def __mul__(self,other):
    new_num = self.num*other.num
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  # division

  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    return '{}/{}'.format(new_num,new_den)
```

```python
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```
