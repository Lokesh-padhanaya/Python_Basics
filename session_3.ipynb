# Sequence sum
1/1! + 2/2! + 3/3! + ....

```python
n = int(input("Enter the n: "))

result = 0
fact = 1

for i in range(1, n+1):
  fact = fact * i
  result = result + i/fact

print(result)
```

# Pattern 1
 *<br>
 **<br>
 ***<br>
 ****<br>

```python
rows = int(input("Enter the number of rows here: "))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print("*", end = "")
  print()

```

# Pattern 2
1<br>
121<br>
12321<br>
1234321<br>

```python
row = int(input("Enter the number of rows here: "))

for i in range(1, row+1):
  for j in range(1, i+1):
    print(j, end = "")
  for k in range(i-1, 0, -1):
    print(k, end = "")
  print()
```

# Loop Control Statement
- Break
- Continue
- Pass

```python
# Break : -> break is used to break the loop
# print the prime numbers between two numbers

lower = int(input("Enter the lower limit: "))
upper = int(input("Enter the upper limit: "))

for i in range (lower , upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i, end = ",")
```

###### Continue : -> continue skips the coditional iteration and move to the next iteration without executing the code next to the continue statement.

```python
# Continue
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
# Pass : -> when the code is not complete but we want machine to not throw the error for that incomplete code.

for i in range(1,10):
  pass
```

## Creating Strings

```python
s = 'hello'
print(s)

# To handle it's raining outside. type of statement because it's already has one ' and machine get confuse with this.
s = "hello"
print(s)

# For multiline strings
s = '''hello'''
print(s)

s = """hello"""
print(s)

s = str('hello')
print(s)
```

## Accessing Substrings from String

```python
# Indexing

# Positive Indexing
s = 'hello world'
print(s[0])

print(s[4])
```

```python
# Negative Indexing
print(s[-1])
```

```python
# Slicing

s = 'hello world'
print(s[0:5])

print(s[0:])

print(s[:5])

print(s[0:6:2])
```

```python
# Negative slicing

s = 'hello world'
print(s[6:0:-1])
```

```python
# reverse the string

s = 'hello world'
print(s[::-1])
```

```python
s = 'hello world'
print(s[-5:])
```

```python
s = 'hello world'
print(s[-1:-6:-1])
```

Editing and Deleting in Strings

```python
s = 'hello world'
s[0] = 'H'

# Python strings are immutable. when we apply functions, new string is generated after appling that function but the original string remains same.
```

```python
s = 'hello world'
del s
print(s)
# becuse complete s has deleted.
```
