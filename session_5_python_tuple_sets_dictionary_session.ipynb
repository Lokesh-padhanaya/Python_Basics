# Tuples

A tuple in python is similar to a list. The difference between the two is that we can not change the elements if a tuple once it is assigned we can change the elements of a list.

In short, a tuple is an immutable list. A list can not be changed in any way once it is created.

Characterstics:
- Ordered
- Unchangeble
- Allows duplicate

# Plan of attack
- Creating a Tuple
- Accessing items
- Editing items
- Adding items
- Deleting items
- Operations on Tuples
- Tuple Functions

# Creating Tuples

```python
# Empty
t1 = ()
print(t1)


# Create a tuple with a single item
t2 = ("hello")
print(t2)
print(type(t2))
# for single item we can't create tuple as usually we do for more than two items,

# Therefore

t2 = ('hello',)
print(t2)
print(type(t2))

#Homogenous Tuple
t3 = (1,2,3,4)
print(t3)

# Hetrogenous Tuple
t4 = (1,2,3,True, [1,2,3])
print(t4)

# 2D tuple (tuple within a tuple)
t5 = (1,2,3,(4,5))
print(t5)


# using type conversion
t6 = tuple('hello')
print(t6)
```

# Difference between Lists and Tuples
- Syntax **([] vs ())**
- Mutability **(list mutable while tuple is not))**
- Speed **(tuple > list)**
- Memomy
- Built in Functionality **(list has more functions)**
- Error **(list is more error prone)**
- Usability **(because tuple is immutable so it will be used where changes are not frequent, while list is used where changes are frequent.)**
- Both are ordered**( means order of the items matters.)**

```python
#error
#list

a = [1,2,3]
b = a
a.append(4)
print(a)
print(b)

# change in b also took place while we made change in a only.
```

```python
# tuple

a = (1,2,3)
b = a
a = a + (4,)
print(a)
print(b)
```

# Special syntax

```python
a,b = (1,2,3)
print(a,b)

# will get error " too many values to unpack (expected 2)"
```

### Value swaping


```python
a = 1
b = 2
a,b = b,a
print(a,b)
```

```python
a,b,*others = (1,2,3,4)
print(a,b)
print(others)
```

```python
# zipping tuples
a = (1,2,3,4)
b = (5,6,7,8)

print(list(zip(a,b)))
print(tuple(zip(a,b)))
```

# Sets
A set is an unordered collection of items. Every set element is unique(no duplicates) and must be immutable(cannot be changed). However, a set itself is mutable. We can add or remove items from it.
Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- can't contain mutable data types

## Creating Sets


```python
# empty
# we can create an empty set like this
# s = {}, because python take it as empty dictionary

s = set()
print(s)
print(type(s))

# 1D and 2D
s1 = {1,2,3}
print(s1)

#s2 = {1,2,4,{4,5}}
#print(s2)

# hetrogenous set
s3 = {1,4.5,'hello', True}
print(s3)
# output only {1, 4.5, 'hello'} because python takes True as 1 and set does not allow duplicates
```

```python
# using type conversion

s4 = set([1,2,4])
print(s4)

# duplicates not allowed
s5 = {1,1,2,2,3,3}
print(s5)

# set can't have mutable items
s6 = {1,2,[3,4]}
print(s6)
```

```python
# Unordered

s1 = {1,2,3}
s2 = {3,2,1}
print(s1 == s2)
```

```python
# Accessing Items

# Indexing (positive + negative) not possible
# slicing not possible
s1 = {1,2,3,4}
#s1[0]

# we can't accessing the items in set, because set is unordered.

```

```python
# Editing Items

s1 = {1,2,3,4}
# s1[0] = 100  not possible
```

# Set Operation
- Union( | )
- Intersection (&)
- Difference(-)
- Symmetric Difference (^)
- Membership Test
- Iteration



```python
#Union( | )

s1 = {1,2,3,4,5}
s2 = {5,6,7,8}
print(s1 | s2)
```

```python
# Intersection(&)
s1 = {1,2,3,4,5}
s2 = {5,6,7,8}

print(s1 & s2)
```

```python
# Difference ( - )
s1 = {1,2,3,4,5}
s2 = {5,6,7,8}

print(s1 - s2)
```

```python
# Symmetric Difference ( ^ ) -> print everything except common
s1 = {1,2,3,4,5}
s2 = {5,6,7,8}
print(s1 ^ s2)
```

```python
# Membership Test
print(1 in s1)

print(1 not in s1)
```

# Set functions

```python
#len/sum/min/max/sorted

s1 = {1,2,3,4,5}
print(len(s1))

print(sum(s1))

print(min(s1))

print(max(s1))

sorted(s1)
```

```python
# Union / update
s1 = {1,2,3,4}
s2 = {4,5,6,7}

# s1 | s2
s1.union(s2)   # both gives same result

# update
s1.update(s2)  # Update function do the union and brings a permanent change on s1 and leave the s2 unchanged.
print(s1)
print(s2)
```

```python
# Intersection / intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)


s1.intersection_update(s2)  #intersection_update function do the intersection and brings a permanent change on s1 and leave the s2 unchanged.
print(s1)
print(s2)
```

```python
# difference / difference_update

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.difference(s2)

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference / symmetric_difference_update

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint / issubset / issuperset

s1 = {1,2,3,4}
s2 = {5,6,7,8}

s1.isdisjoint(s2)
```

```python
#copy
s1 = {1,2,3,4}
s2 = s1.copy()
print(s1)
print(s2)
```

# Dictionary
Dictionary in Python is a collection of keys and values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = {'name' : 'Lokesh','age': '27','gender': 'male'}

Characterstics:
- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items

```python
# Empty dictionary
d = {}
d

# 1D dictionary
d1 =  {'name':'Lokesh','gender':'male'}
d1

# with mixed keys
d2 = {(1,2,3): 1,'hello':'world'}
d2

# 2D dictionary  -> uses in JSON
d3 = {
    'name':'Lokesh',
    'college':'FAU',
    'Sem':'2',
    'subjects':{
        'dsa':'50',
        'maths':'50',
        'english':'49'
    }
}
d3
```

```python
# Using sequence and dict function
d4 = dict([('name','Lokesh'),('age',27),('gender','male')])
d4
```

```python
# can't have duplicate keys
d5 = {'name':'Lokesh','name':'Sachin'}
d5
```

```python
# mutable items as keys
d6 = {'name':'Lokesh',[1,2,3]:2}
d6
```

```python
# Accessing Items
my_dict = {'name':'Lokesh','age':'27'}
# my_dict[0]  -> not allowed
my_dict['name']  # ->  allowed

# get
my_dict.get('name')
```

# Dictionary Comprehension
 {***key*** : ***value*** for ***vars*** in ***iterable***}

```python
# print 1st 10 numbers and their squares
{i: i**2 for i in range(1,11)}
```

```python
# using existing dict
distances = {'delhi': 1000, 'mumbai':2000,'banglore':3000}

{key:value *0.62 for (key,value) in distances.items()}
```

```python
# using zip
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{i:j for (i,j) in zip(days,temp_C)}
```

```python
# using if condition
products = {'phone':10, 'laptop':0,'charger':32,'tablet':0}

# we dont want to print the keys whose value = 0

{key:value for (key,value) in products.items() if value > 0}
```

```python
# nested comprehension
{
    2:{1:2,2:4,3:6,4:8},
    3:{1:3,2:6,3:9,4:12},
    4:{1:4,2:8,3:12,4:16}
 }



# print tables of numbers from 2 to 4

{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
